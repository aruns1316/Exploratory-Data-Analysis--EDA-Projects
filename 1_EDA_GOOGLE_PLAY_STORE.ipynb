```python
# 1MB = 1024kb.

def size_process(item):
  if str(item)[-1] == 'M':
    result =  float(str(item).replace('M', '')) # Indent this line to be part of the 'if' block
    result = result * 1024
  elif str(item)[-1] == 'k':
    result = float(str(item).replace('k',''))
    return result
  else:
    return (np.NaN) # You might want to handle the case where the input is neither in MB nor kB
```

```python
df_copy['Size'].apply(size_process)
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_process)
```
